###0. Import dependencies

```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
import plotly.graph_objects as go
import numpy as np

# Load the diabetes dataset
data = load_diabetes()
```

### 1. Creating variables and date

```python
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='target')
```

### 2. Splitting the data - training and test

```python
# Step 2: Splitting the data into training and testing sets (80% train, 20% test)
# The function automatically handles the 2D matrix structure of X
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 3. Creating instance of the Ridge class and fit the model into training data

```python
regressor = Ridge(alpha=1.0) # Penaly parameter
regressor.fit(X_train, y_train)
```

### 4. Actual predictions on test data

```python
# Step 4: Making predictions on the test data
y_pred = regressor.predict(X_test)
# print(y_pred)
```

### 5. Plotting the data for visual representation

```python
fig = go.Figure()
# CAMBIO AQUÍ: Convertimos y_test en un array plano sin buscar nombres de columnas
fig.add_trace(go.Scatter(x=np.asarray(y_test).flatten(), y=y_pred.flatten(), mode='markers'))
fig.add_trace(go.Scatter(x=[0, 350], y=[0, 350], mode='lines'))
fig.update_layout(title='Actual vs. Predicted Values',
                  xaxis_title='Actual Values',
                  yaxis_title='Predicted Values')
fig.show()
```
