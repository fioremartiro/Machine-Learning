# Implementation of Polynomial Regression

```python
import numpy as np
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# STEP 1: Generate synthetic non-linear data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 3 * x ** 2 + 2 * x + np.random.normal(0, 10, 100)
```

```python
# STEP 2: Reshape arrays to 2D column vectors for scikit-learn
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
# STEP 3: Split the dataset into Training (80%) and Testing (20%) sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# STEP 4: Create the polynomial "disguise" (adds X^2 column)
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
# STEP 5: Train the Linear Regression model on the polynomial features
model = LinearRegression()
model.fit(x_train_poly, y_train)
```

```python
# STEP 6: Generate a smooth line for the final visualization
x_line = np.linspace(0, 10, 200).reshape(-1, 1)
x_line_poly = poly.transform(x_line)
y_line_pred = model.predict(x_line_poly)
```

Visualize the original data and the polynomial regression curve

```python
# STEP 7: Plot everything using Plotly
trace_train = go.Scatter(x=x_train.flatten(), y=y_train.flatten(), mode='markers', name='Training Data', marker=dict(color='blue'))
trace_test = go.Scatter(x=x_test.flatten(), y=y_test.flatten(), mode='markers', name='Testing Data', marker=dict(color='green'))
trace_regression = go.Scatter(x=x_line.flatten(), y=y_line_pred.flatten(), mode='lines', name='Polynomial Regression', line=dict(color='red', width=3))

layout = go.Layout(
    title='Polynomial Regression (Smooth Curve)',
    xaxis=dict(title='Independent Variable - X'),
    yaxis=dict(title='Dependent Variable - Y')
)

figure = go.Figure(data=[trace_train, trace_test, trace_regression], layout=layout)
figure.show()
```
