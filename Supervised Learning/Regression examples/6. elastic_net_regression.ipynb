### 0. Import dependencies

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet

# Load the diabetes dataset
diabetes_data = load_diabetes()
```

### 1. Creating variables and date

```python
X = pd.DataFrame(diabetes_data.data, columns=diabetes_data.feature_names)
y = pd.Series(diabetes_data.target, name='target')
print ("This is the data:", y.head())
```

### 2. Splitting the data - training and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 3. Creating instance of the Elastic Net class and fit the model into training data

```python
regressor = ElasticNet()

testing = {
    'alpha': [0.001, 0.01, 0.1, 1, 10, 100],
    'l1_ratio': [0.2, 0.5, 0.8]
}

search = GridSearchCV(regressor, testing, scoring='neg_mean_squared_error', cv=5)

search.fit(X_train, y_train)

print("Mejores parámetros encontrados:", search.best_params_)
```

### 4. Actual predictions on test data

```python
y_pred = search.best_estimator_.predict(X_test)
```

### 5. Plotting the data for visual representation

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=y_test, y=y_pred, mode='markers', name='Predictions'))
fig.add_trace(go.Scatter(x=[y_test.min(), y_test.max()], y=[y_test.min(), y_test.max()], mode='lines', name='Perfect Fit', line=dict(color='red')))
fig.update_layout(title='Actual vs Predicted (Elastic Net)', xaxis_title='Actual', yaxis_title='Predicted')
fig.show()
```
